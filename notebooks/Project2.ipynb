#### <font color="darkslategray">The greenhouse gas effect plays an essential role in Earth's steady-state climate. Without greenhouse gases, the Earth would be uninhabitable due to lower temperatures. Greenhouse gases retain heat in the climate system. Visible light, or shortwave radiation, is able to pass more easily through Earth's atmosphere than infrared radiation, or longwave radiation. The infrared radiation that is not able to pass as easily through the atmosphere and that is radiated from Earth's surface is absorbed by greenhouse gases. Greenhouse gases re-emit this heat throughout the climate system. The atmosphere is composed of 78% N₂, 21% O₂, ~1% Ar, and about 0.04% greenhouse gases. The distinction between, say, a greenhouse gas such as CH₄ and a gas such as N₂ is that CH₄ and all other greenhouse gases are triatomic molecules instead of diatomic. Greenhouse gases have more points of vibration that allow them to absorb and emit heat in this way.</font>

#### <font color="darkslategray"> In the calculations above, we see that CH4 has increased the most relative to pre-industrial concentrations by a value of 172.98%. Despite CH4 having the most significant increase, it is not commonly the greenhouse gas we associate with climate change. CO2 is the most discussed greenhouse gas, and this is due to its residence time. Potency and residence time are important factors to consider when discussing greenhouse gases. Potency refers to the amount of heat the gas traps in the atmosphere over a given time. Residence time refers to how long the gas remains in the atmosphere before it is broken down by chemical processes. Potency can also be referred to as a gas’s 'radiative efficiency,' measured in Wm⁻²ppb⁻¹.</font>
 </font>

|Gas|Radiative Efficiency (Wm^-2ppb^-3)|Residence Time (Years)|
|--|--|--|
|CO2|1.33*10^-5|300-1000|
|NH4|5.7*10^-4|10-12|
|N2O|2.8*10^-3|~112|

#### <font color="darkslategray"> From the table above, we can see that N2O and CH4 are both far more potent than CO2. If we then consider residence time, we see that CO2 has a residence time approximately 100 times longer than CH4 and 10 times longer than N2O. It is CO2's long residence time that makes it so alarming when it comes to the discussion of global warming and the consequences of heavy industrialization. We know that CO2 has oscillated between 180 and 280 ppm throughout the 800,000 years before industrialization. This is known from samplings of ice cores and examination of the air bubbles trapped within. These fluctuations were tied to Earth's natural cycles; however, anthropogenic forcings have taken levels to uncharted territory.

```python
import matplotlib.pyplot as plt

#Create a dictionary for the datasets.
datasets = {
    'Berkeley GMT': gmt_berkley,
    'ERA5 GMT': gmt_era5,
    'GISTEMP GMT': gmt_gistemp,
    'HadCRUT5 GMT': gmt_had,
    'JRA-55 GMT': gmt_jra,
    'NOAA GMT': gmt_noaa,
    'Berkeley LSAT': lsat_berkley,
    'CRUTEM5 LSAT': lsat_crutem,
    'ERSSTv5 SST': sst_ers,
    'HadSST3 SST': sst_had3,
    'HadSST4 SST': sst_had4,
}


plt.figure(figsize=(14, 7))

#Create a for, in function to derive the necessary values from the datasets
for name, data in datasets.items():
    # Make sure the column names don't have extra spaces to avoid the Key error when calling the column names. 
    data.columns = data.columns.str.strip()
    # Find the temperature column by looking for "(degC)" in its name
    temp_columns = []
    for col in data.columns:
        if '(degC)' in col:
            temp_columns.append(col)
    #Assign the name of the temperature columns to temp_col
    temp_col = temp_columns[0]  
    
    #ISolate the rows between 2012 and 2022
    filtered_data = data[(data['Year'] >= 2012) & (data['Year'] <= 2022)]
    
    
    plt.plot(filtered_data['Year'], filtered_data[temp_col], label=name)


plt.title("Global, Land, and Ocean Surface Temperature Anamolies (2012–2022)")
plt.xlabel("Year")
plt.ylabel("Temperature Anamoly(°C)")

# Create a legend that is outside of the plot 
plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))


plt.grid(True)

plt.show()



```
